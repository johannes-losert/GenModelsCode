```python
def print_extractor_results(extractor) -> None:
    """
    Print the first 5 results from a credit card extractor.
    
    Args:
        extractor: Instance of a CreditCardExtractor
    """
    print(f"\n{'-'*20} {extractor.site_name} Results {'-'*20}")
    try:
        results = extractor.extract()
        print(f"Total cards found: {len(results)}\n")
        print("First 5 card descriptions:\n")
        
        for i, card in enumerate(results[:20], 1):
            print(f"Card {i}:")
            description_lines = card.description.split(". ")
            formatted_description = "\n".join(f"  {line}." for line in description_lines if line)
            print(formatted_description)
            print()
            
    except Exception as e:
        print(f"Error processing {extractor.site_name}: {str(e)}")
        print("Make sure you have the correct HTML file in the directory.")

def main():
    # Test each extractor
    extractors = [
        CardRatingsExtractor(),
        CreditKarmaExtractor(),
        USNewsExtractor()
    ]
    
    for extractor in extractors:
        print_extractor_results(extractor)

if __name__ == "__main__":
    main()
```
